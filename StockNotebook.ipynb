```python
import pandas as pd

data = pd.read_csv('BSE.csv')
data.head()
```

```python
def create_df(data):
    rows = []
    for i in data:
        rows.append([i['instrument'], i['name'], i['symbol'], i['price']])
    
    df = pd.DataFrame(rows, columns=['instrument', 'name','symbol', 'price'])
    return df



items = []
max_price = 15
for index, row in data.iterrows():
    instrument = row['instrument_key']
    price = row['last_price']
    symbol = row['tradingsymbol']
    name = row['name']
    if price>0 and price<max_price:
        newRow = {"instrument":instrument, "price":price, "symbol":symbol,"name":name}
        items.append(newRow)
        
stockdf = create_df(items)
stockdf.head(10)
```
